## Filter The Data (1 for spam)

```python
# Extract the numeric part of "Email No." and convert it to integer
df['Email No. Numeric'] = df['Email No.'].str.extract('(\d+)').astype(int)
filtered_data = df[(df['Email No. Numeric'] > 100) & (df['Email No. Numeric'] < 200) & (df['Prediction'] == 1)]
filtered_data_selection = filtered_data[['Email No.', 'Prediction']]
print(filtered_data_selection)
```

## Use Group By to Explore Data
Group by a specific column and perform an aggregation operation. For example, group by the column 'Category' and find the mean of the column 'Rating'.

```python
# Filter the DataFrame to include only rows where the count of 'the' is greater than 10
filtered_df = df[df['the'] > 10]

# Group the filtered DataFrame by 'Prediction' and calculate the mean of the 'the' column
grouped_data = filtered_df.groupby('Prediction')['the'].mean()

print(grouped_data)
```

## ML Part Of this Data Analysis

```python
from sklearn.model_selection import train_test_split

# Assuming the last column is the label for spam or not spam
X = df.drop(columns=['Email No.', df.columns[-1]])
y = df[df.columns[-1]]

# Splitting the dataset into the Training set and Test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Initialize the models
decision_tree = DecisionTreeClassifier(random_state=0)
random_forest = RandomForestClassifier(random_state=0)

# Train the Decision Tree model
decision_tree.fit(X_train, y_train)
# Train the Random Forest model
random_forest.fit(X_train, y_train)

# Make predictions with both models
dt_predictions = decision_tree.predict(X_test)
rf_predictions = random_forest.predict(X_test)

# Calculate accuracy
dt_accuracy = accuracy_score(y_test, dt_predictions)
rf_accuracy = accuracy_score(y_test, rf_predictions)

print(f"Decision Tree Accuracy: {dt_accuracy}")
print(f"Random Forest Accuracy: {rf_accuracy}")
```

## Analyzing Feature Importance
The Random Forest model can provide insights into which features (in this case, words) are most important when it comes to predicting spam. This is useful for identifying which words are strong indicators of spam

```python
feature_importances = random_forest.feature_importances_
# Get the indices of the 50 most important features
most_important_indices = feature_importances.argsort()[-50:]
# Map these indices back to the column names in your DataFrame to get the words
important_words = X.columns[most_important_indices]
print(important_words)
```

## Identify Misclassified Emails
Compare the predictions to the actual labels to identify which emails were misclassified.

```python
# Assuming the model is already trained and named `random_forest`
predictions = random_forest.predict(X_test)
# Create a DataFrame for analyzing misclassifications
misclassified_df = X_test.copy()
misclassified_df['Actual Label'] = y_test
misclassified_df['Predicted Label'] = predictions

# Identify misclassified emails
misclassified_df = misclassified_df[misclassified_df['Actual Label'] != misclassified_df['Predicted Label']]
```

## Analyze Misclassified Emails
Now that I have a dataset of misclassified emails, I can analyze them to see if there are common words or features that might be leading to these errors. This could involve looking at the most common words in misclassified emails or comparing their word frequencies to those in correctly classified emails.

```python
# This is a conceptual example to guide the thought process
# Count the occurrence of each word in misclassified emails
word_counts = misclassified_df.drop(columns=['Actual Label', 'Predicted Label']).sum(axis=0)

# Sort the words by their count to see the most frequent ones
most_common_words_in_misclassified = word_counts.sort_values(ascending=False).head(50)
print(most_common_words_in_misclassified)
```
